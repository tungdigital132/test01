IMPORT Broker or DP

```python
import pyodbc

# Kết nối tới SQL Server
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=.\\MSSQLSERVER02;"
    "DATABASE=DatabaseSymbols;"
    "UID=sa01;"
    "PWD=sa123456"
)

cursor = conn.cursor()

# Câu lệnh chèn hoặc cập nhật dữ liệu vào bảng
insert_or_update_query = """
IF EXISTS (
    SELECT 1 FROM [Broker or DP] WHERE [Broker or DP number] = ?
)
BEGIN
    UPDATE [Broker or DP]
    SET [Broker or DP name] = ?, [Broker or DP ID] = ?
    WHERE [Broker or DP number] = ?
END
ELSE
BEGIN
    INSERT INTO [Broker or DP] (
        [Broker or DP number],
        [Broker or DP name],
        [Broker or DP ID]
    ) VALUES (?, ?, ?)
END
"""

# Dữ liệu để chèn
data = [
    ('01', 'Tradingview - FXCM', 'TDVF'),
    ('02', 'Tradingview - TVC', 'TDVT'),
    ('03', 'Tradingview - OANDA', 'TDVO'),
    ('04', 'Tradingview - PEPERSTONE', 'TDVP'),
    ('05', 'MT5 - FPMarket', 'MT5F'),
    ('06', 'MT5 - OANDA', 'MT5O')
]

try:
    # Chèn hoặc cập nhật dữ liệu từng dòng
    for record in data:
        cursor.execute(
            insert_or_update_query,
            record[0],  # Kiểm tra tồn tại (Broker or DP number)
            record[1],  # Giá trị cập nhật (Broker or DP name)
            record[2],  # Giá trị cập nhật (Broker or DP ID)
            record[0],  # Giá trị WHERE cho UPDATE
            record[0], record[1], record[2]  # Giá trị để chèn nếu không tồn tại
        )

    # Lưu thay đổi
    conn.commit()
    print("Dữ liệu đã được chèn hoặc cập nhật thành công.")
except Exception as e:
    print("Lỗi:", e)
finally:
    # Đóng kết nối
    cursor.close()
    conn.close()
```

GỘP Data Table

```python
import pyodbc

# Kết nối tới SQL Server
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=.\\MSSQLSERVER02;"
    "DATABASE=DatabaseSymbols;"
    "UID=sa01;"
    "PWD=sa123456"
)

cursor = conn.cursor()

# Truy vấn gộp dữ liệu từ các bảng vào Data Table và thêm sắp xếp
merge_query = """
INSERT INTO [Data Table] (
    [IIT ID Number],
    [IIT ID Name],
    [Broker or DP number],
    [Broker or DP name],
    [Broker or DP ID],
    [Instrument Reference Name],
    [Instrument Reference ID],
    [Instrument Type Name],
    [Instrument Type ID],
    [Specific Instrument Type Name],
    [Specific Instrument SYMBOL]
)
SELECT 
    CONCAT(
        r.[Instrument Type ID], 
        r.[Instrument Reference ID], 
        CAST(t.[Broker or DP number] AS VARCHAR(2))
    ) AS [IIT ID Number],
    CONCAT(
        r.[Instrument Reference Name], '.', b.[Broker or DP ID]
    ) AS [IIT ID Name],
    CAST(t.[Broker or DP number] AS VARCHAR(2)) AS [Broker or DP number],
    b.[Broker or DP name],
    b.[Broker or DP ID],
    r.[Instrument Reference Name],
    r.[Instrument Reference ID],
    r.[Instrument Type Name],
    r.[Instrument Type ID],
    t.[Specific Instrument Type Name],
    t.[Specific Instrument SYMBOL]
FROM [Reference Name] r
LEFT JOIN [Specific SYMBOL - Tradingview] t
    ON r.[Instrument Reference ID] = t.[Instrument Specific ID]
LEFT JOIN [Broker or DP] b
    ON CAST(t.[Broker or DP number] AS VARCHAR(2)) = CAST(b.[Broker or DP number] AS VARCHAR(2))
UNION ALL
SELECT 
    CONCAT(
        r.[Instrument Type ID], 
        r.[Instrument Reference ID], 
        CAST(f.[Broker or DP number] AS VARCHAR(2))
    ) AS [IIT ID Number],
    CONCAT(
        r.[Instrument Reference Name], '.', b.[Broker or DP ID]
    ) AS [IIT ID Name],
    CAST(f.[Broker or DP number] AS VARCHAR(2)) AS [Broker or DP number],
    b.[Broker or DP name],
    b.[Broker or DP ID],
    r.[Instrument Reference Name],
    r.[Instrument Reference ID],
    r.[Instrument Type Name],
    r.[Instrument Type ID],
    f.[Specific Instrument Type Name],
    f.[Specific Instrument SYMBOL]
FROM [Reference Name] r
LEFT JOIN [Specific SYMBOL - FPMarket] f
    ON r.[Instrument Reference ID] = f.[Instrument Specific ID]
LEFT JOIN [Broker or DP] b
    ON CAST(f.[Broker or DP number] AS VARCHAR(2)) = CAST(b.[Broker or DP number] AS VARCHAR(2))
ORDER BY 
    [Broker or DP number] ASC, 
    [Instrument Reference ID] ASC;
"""

try:
    # Thực thi truy vấn
    cursor.execute(merge_query)

    # Lưu thay đổi
    conn.commit()
    print("Dữ liệu đã được gộp và sắp xếp thành công.")
except Exception as e:
    print("Lỗi:", e)
finally:
    # Đóng kết nối
    cursor.close()
    conn.close()
```
