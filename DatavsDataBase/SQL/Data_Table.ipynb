```python
import pyodbc

# Thông tin kết nối tới SQL Server
server = '.\\MSSQLSERVER02'
database = 'DatabaseSymbols'
username = 'sa01'
password = 'sa123456'

# Chuỗi kết nối
connection_string = f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}"

try:
    # Kết nối tới SQL Server
    connection = pyodbc.connect(connection_string)

    # Danh sách IIT ID Number
    iit_ids = [10201,10301,10401,10501,10601,10701,10801,11001,35603,48104,10205,10305,10405,10505,10605,10705,10805,11005,35605,48105]

    

    # Tạo câu lệnh SQL
    query = f"""
    SELECT * 
    FROM [Data Table]
    WHERE [IIT ID Number] IN ({','.join('?' for _ in iit_ids)});
    """

    # Thực thi câu lệnh SQL
    cursor = connection.cursor()
    cursor.execute(query, iit_ids)

    # Lấy tên cột
    columns = [column[0] for column in cursor.description]

    # Xác định độ rộng tối đa cho mỗi cột
    column_widths = [max(len(col), 15) for col in columns]  # Đặt độ rộng tối thiểu là 15
    for row in cursor.fetchall():
        for i, value in enumerate(row):
            column_widths[i] = max(column_widths[i], len(str(value)))

    # In tiêu đề cột
    header = " | ".join(col.ljust(column_widths[i]) for i, col in enumerate(columns))
    print(header)
    print("-" * len(header))

    # In dữ liệu
    cursor.execute(query, iit_ids)  # Thực thi lại vì con trỏ đã duyệt hết dữ liệu
    rows = cursor.fetchall()
    for row in rows:
        row_data = " | ".join(str(value).ljust(column_widths[i]) for i, value in enumerate(row))
        print(row_data)

except pyodbc.Error as e:
    print("Lỗi kết nối hoặc truy vấn:", e)

finally:
    # Đóng kết nối
    if 'connection' in locals():
        connection.close()
```
